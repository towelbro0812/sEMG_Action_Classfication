```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.metrics import f1_score, confusion_matrix, precision_recall_curve, roc_curve
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler


import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

import warnings
warnings.filterwarnings(action='ignore')
```

```python
from pycaret.classification import ClassificationExperiment
# 初始化
data = pd.read_csv("merged_data.csv")
exp1 = ClassificationExperiment()
exp1.setup(data=data, target='Action', session_id=123)
```

```python
best_model = exp1.compare_models(
    sort='Accuracy',
    n_select=5,
    exclude=['dummy', 'svm', 'ada','ridge','qda','nb','lda','lr']
)
```

```python
exp1.plot_model(best_model[0], plot='confusion_matrix')
```

```python
exp1.plot_model(best_model[0], plot='auc')
```
